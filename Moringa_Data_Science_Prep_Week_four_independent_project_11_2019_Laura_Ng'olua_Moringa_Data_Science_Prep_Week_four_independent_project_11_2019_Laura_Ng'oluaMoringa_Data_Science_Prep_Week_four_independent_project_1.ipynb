## Loading datasets

```
#importing pandas and numpy libraries
import pandas as pd
import numpy as np
```

```
#loading autolib dataset
auto= 'http://bit.ly/Autolibdataset'

#reading the dataset and storing it in a desired variable
autolib=pd.read_csv(auto)
#autolib
#viewing only the first ten entries
autolib.head(10)
```

## Identifying null values(Validity)

```
#calculate the null values
np.count_nonzero(autolib.isnull())
```

```
#calculate the null values in the displayed comment and scheduled at columns
np.count_nonzero(autolib['Displayed comment'].isnull())
```

```
#calculate the null values in the scheduled at columns
np.count_nonzero(autolib['Scheduled at'].isnull())
```

```
#dropping the display comment column and scheduled at columns
autolib.drop(["Displayed comment", "Scheduled at"], axis = 1, inplace = True)

autolib.head(10)
```

```
np.count_nonzero(autolib.isnull())
```

```
#deleting entries with at least ninety percent nan

# Only drop columns which have at least 90% non-NaNs
autolib.dropna(thresh=int(autolib.shape[0] * .9), axis=1)
```

# question one

```
#What is the most popular hour for returning cars?
auto_clean['BlueCar Diff']= auto_clean.sort_values(['Collection Timestamp']).groupby(['ID'])['Bluecar counter'].apply(lambda x:  x.shift(1) - x)
auto_clean
```

```
print(auto_clean[(auto_clean['City']=='Paris') & (auto_clean['BlueCar Diff'] > 0)].groupby(['hour'])['BlueCar Diff'].sum().sort_values(ascending =True).head(10))
```

# question two

```
#What is the most popular hour for returning cars?
auto_clean.groupby(['hour'])['Cars'].sum().diff().sort_values(ascending =False).head(3)
```

# question three

```
#What station is the most popular?
#Overall?
auto_clean.groupby('ID')['BlueCar Diff'].max().sort_values(ascending= False).head(10)
```

```
#What station is the most popular?
#At the most popular picking hour?
print(auto_clean[(auto_clean['hour']==4) & (auto_clean['BlueCar Diff'] > 0)].groupby(['ID'])['BlueCar Diff'].sum().sort_values(ascending =True).head(10))
```

# question four

```
#What postal code is the most popular for picking up Bluecars? 
#Does the most popular station belong to that postal code?
#Overall?
auto_clean.groupby('Postal code')['BlueCar Diff'].max().sort_values(ascending= False).head(10)
```

```
#What postal code is the most popular for picking up Bluecars? 
#Does the most popular station belong to that postal code?
#at the most popular picking hours
print(auto_clean[(auto_clean['hour']==4) & (auto_clean['BlueCar Diff'] > 0)].groupby(['Postal code'])['BlueCar Diff'].sum().sort_values(ascending =True).head(10))
```
